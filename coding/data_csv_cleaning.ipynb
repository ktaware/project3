```python
# Import dependencies.
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
from datetime import datetime as dt
import json
```

### Extract the airports.csv Data

```python
# Read the data into a Pandas DataFrame.
airports_df = pd.read_csv('../data/raw_data/airports.csv')
airports_df.head()
```

```python
# Get a brief summary of the airports_df DataFrame.
airports_df.info()
```

```python
# Get the airports_df columns.
airports_df.columns
```

```python
# Drop unwanted ICAO, Timezone, DST, Timezone.1 columns..
airports_cleaned = airports_df.drop(columns=['ICAO', 'Timezone', 'DST', 'Timezone.1'])
# Rename the Name, City, Country, IATA, Latitude, Longitude, Altitude columns.
airports_cleaned = airports_cleaned.rename(columns={'Name': 'name',
                                          'City': 'city', 
                                          'Country': 'country', 
                                          'IATA': 'airport_id', 
                                          'Latitude': 'lat', 
                                          'Longitude': 'lng',
                                          'Altitude': 'alt'
                                          })
airports_cleaned.head()
```

```python
# Replace '\N' string entries, drop all NaN values and reindex DataFrame
airports_final = airports_cleaned.replace(r'^\\N$', np.nan, regex=True).dropna().reset_index(drop=True, inplace=False)
airports_final.head()
```

```python
# Get an updated summary of the airports_final DataFrame.
airports_final.info()
```

```python
# Export airports_final as CSV files.
airports_final.to_csv("../data/airports_cleaned.csv", index=False)
```

### Extract the airlines.csv Data

```python
# Read the data into a Pandas DataFrame.
airlines_df = pd.read_csv('../data/raw_data/airlines.csv')
airlines_df.head()
```

```python
# Get a brief summary of the airlines_df DataFrame.
airlines_df.info()
```

```python
# Get the airlines_df columns.
airlines_df.columns
```

```python
# Drop unwanted ICAO, Callsign columns.
airlines_cleaned = airlines_df.drop(columns=['ICAO', 'Callsign'])

# Rename the Name, IATA, Country, Acive columns.
airlines_cleaned = airlines_cleaned.rename(columns={'Name': 'name',
                                          'IATA': 'airline_id', 
                                          'Country': 'origin_country', 
                                          'Active': 'active'                                          
                                          })
airlines_cleaned.head()
```

```python
# Drop any duplicate airline_ids and NaN values
airlines_dropped = airlines_cleaned.drop_duplicates(subset=['airline_id']).dropna()

# Reset index for resulting airlines_final DataFrame.
airlines_final = airlines_dropped.reset_index(drop=True, inplace=False)
airlines_final.head()
```

```python
# Export airlines_final as CSV file.
airlines_final.to_csv("../data/airlines_cleaned.csv", index=False)
```

### Extract the routes.csv Data

```python
# Read the data into a Pandas DataFrame
routes_df = pd.read_csv('../data/raw_data/routes.csv')
routes_df.head()
```

### Extract the International_Report_Departures.csv Data

```python
# Read the data into a Pandas DataFrame.
all_flights_df = pd.read_csv('../data/raw_data/International_Report_Departures.csv')
all_flights_df.head()
```

```python
# Get a brief summary of the all_flights_df DataFrame.
all_flights_df.info()
```

```python
# Get the all_flights_df columns.
all_flights_df.columns
```

```python
# Drop unwanted dep_time, dep_delay, arr_time, arr_delay, time_hours columns.
all_flights_dropped = all_flights_df.drop(columns=['data_dte', 'Month',
       'usg_apt_id', 'usg_wac', 'fg_apt_id', 'fg_wac', 'type'])

# Keep only data for two most recent years (2020, 2019)
all_flights_reduced = all_flights_dropped[all_flights_dropped.Year.isin([2020, 2019])]

# Keep only data where a major NYC airport (EWR, JFK, LGA) is the departure point
all_flights_airports_reduced = all_flights_reduced[all_flights_reduced.usg_apt.isin(['EWR', 'JFK', 'LGA'])]

# Remove airlines irrelevant to scope
all_flights_airlines_reduced = all_flights_airports_reduced[all_flights_airports_reduced['carrier'].str.len() <= 2]
all_flights_airlines_reduced = all_flights_airlines_reduced[~all_flights_reduced.carrier.isin(['0Q', 'WI'])]

# Drop any NaN values
all_flights_airlines_reduced = all_flights_airlines_reduced.dropna()
all_flights_airlines_reduced.head()
```

```python
# Rename the Year, usg_apt, fg_apt, airlineid, carrier, carriergroup, Scheduled, Charter and Total columns.
all_flights_cleaned = all_flights_airlines_reduced.rename(columns={'Year': 'year',
                                                'usg_apt': 'dep_airport', 
                                                'fg_apt': 'des_airport', 
                                                'airlineid': 'alid',
                                                'carrier': 'airline_id',
                                                'carriergroup': 'dom_fgn',
                                                'Scheduled': 'scheduled',
                                                'Charter': 'chartered',
                                                'Total': 'total'                                                
                                          })

# Reset index for resulting airlines_final DataFrame.
all_flights_final = all_flights_cleaned.reset_index(drop=True, inplace=False)
all_flights_final = all_flights_final.reset_index(drop=False, inplace=False)
all_flights_final.head()
```

```python
# Export all_flights_final as CSV file.
all_flights_final.to_csv("../data/all_flights_cleaned.csv", index=False)
```
